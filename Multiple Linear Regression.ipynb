## Using Gradient Descent


```python
xi=data.iloc[:,[0,1,2]].values
yi=data.iloc[:,-1].values
```

```python
def gradientdescent(b0,b1,b2,b3,yi,xi,tolerance,eta,verbose=False):
    mag=100
    while mag>tolerance:
        coef=np.array([b0,b1,b2,b3])
        error=yi-(b0+(b1*xi[:,0])+(b2*xi[:,1])+(b3*xi[:,2]))
        derv=np.array([np.sum(error),np.sum(error*xi[:,0]),np.sum(error*xi[:,1]),np.sum(error*xi[:,2])])
        new_co=coef+((2*eta)*derv)
        if verbose==True:
            print(b0,b1,b2,b3)
        b0,b1,b2,b3=new_co[0],new_co[1],new_co[2],new_co[3]
        mag=np.sqrt(np.sum(error)**2+np.sum(error*xi[:,0])**2+np.sum(error*xi[:,1])**2+np.sum(error*xi[:,2])**2)
    return b0,b1,b2,b3
```

```python
gradientdescent(0,0,0,0,yi,xi,0.01,0.001,verbose=True)
```

# Coordinate Descent

```python
xi=data.iloc[:,[0,1,2]].values
yi=data.iloc[:,-1].values
```

```python
def coordinate_descent(X,y,alpha = .03, num_iters=20,normalize=False,verbose=False):
    if normalize==False:
        X=np.c_[np.ones(X.shape[0]),X]
        m,n = X.shape
        theta = np.zeros(n).reshape(-1,1)
        print(theta.flatten())
    
        for i in range(num_iters):
            for j in range(n):
                h = X @ theta 
                gradient =  (X[:,j] @ (h-y))
                theta[j] = theta[j] - alpha * gradient
                if verbose==True:
                    print(theta.flatten())
                       
        return theta 
    elif normalize==True:
        X=np.c_[np.ones(X.shape[0]),X]
        X = X / (np.linalg.norm(X,axis = 0)) 
        m,n = X.shape
        theta = np.zeros(n).reshape(-1,1)
        print(theta.flatten())
        for i in range(num_iters):
            for j in range(n):
                h = (X[:,0:j] @ theta[0:j]) + (X[:,j+1:] @ theta[j+1:])
                theta[j] =  (X[:,j] @ (y - h))
                if verbose==True:
                    print(theta.flatten())
                       
    return theta
```

```python
coordinate_descent(xi,yi.reshape(-1,1),normalize=True)
```
